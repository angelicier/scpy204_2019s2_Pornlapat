```python
from astroquery.sdss import SDSS
query = "SELECT s.z, s.zErr, p.petroMag_g, p.petroMagErr_g FROM  photoObjAll p JOIN specObjAll s ON s.bestObjID = p.ObjID WHERE s.class = 'qso' and s.z between 0.3 and 0.4 and s.zWarning = 0 and s.zErr <= 0.01 and p.petroMagErr_g <= 0.99"
w = SDSS.query_sql(query)
print(w)
```

```python
z = w['z']
zErr = w['zErr']
```

```python
petroMag_g = w['petroMag_g']
petroMagErr_g = w['petroMagErr_g']
```

```python
x = z
x_err = zErr<=1
y = petroMag_g
y_err = petroMagErr_g<=1

figure = plt.figure(0)
axes = plt.gca()
axes.set_xlim([0.3,0.4])
axes.set_ylim([10,30])
plt.errorbar(x, y, xerr=zErr, yerr=petroMagErr_g, fmt='o', markersize=0.5, alpha=0.5, elinewidth=0.5, color='lightseagreen')
plt.title('Relation between PetroMag and z', fontdict=None, loc='center', )
plt.ylabel('PetroMag')
plt.xlabel('z')
plt.show()
```
